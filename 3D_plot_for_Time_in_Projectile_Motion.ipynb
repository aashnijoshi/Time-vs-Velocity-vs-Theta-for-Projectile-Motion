```python

#Importing different libraries
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
from matplotlib import pyplot as plt
import math
import numpy as np

#Giving the user an option to choose a planet they want to project the projectile on
print("What planet are you on?")
print("Please press 1 for Mercury")
print("Please press 2 for Venus")
print("Please press 3 for Earth")
print("Please press 3.1 for Earth's Moon")
print("Please press 4 for Mars")
print("PLEASE PRESS 4.1 FOR MARS' FIRST MOON - PHOBOS")
print("Please press 5 for Jupiter")
print("Please press 5.1 for JUPITER'S MOON - EUROPA")
print("Please press 5.2 for JUPITER'S MOON - GANYMEDE")
print("Please press 6 for Saturn")
print("Please press 6.1 for Saturn's moon - Titan")
print("Please press 6.2 for Saturn's moon - Enceladus")
print("Please press 7 for Uranus")
print("Please press 7.1 for Uranus' moon - Oberon")
print("Please press 7.2 for Uranus' moon - Ariel")
print("Please press 8 for Neptune")
print("If you still believe Pluto is a planet, press 9")
print("If you are hard-wired to survive on the sun, press 0")
planet=input()
pl=float(planet)
g=0

#Substituting g for every planet/moon for its particular acceleration due to gravity
if pl==0:
    g=274
elif pl==1:
    g=3.7
elif pl==2:
    g=8.87
elif pl==3:
    g=9.8
elif pl==3.1:
    g=1.62
elif pl==4:
    g=3.721
elif pl==4.1:
    g=0.0057
elif pl==5:
    g=24.79
elif pl==5.1:
    g=1.315
elif pl==5.2:
    g=1.428
elif pl==6:
    g=10.44
elif pl==6.1:
    g=1.352
elif pl==6.2:
    g=0.133
elif pl==7:
    g=8.87
elif pl==7.1:
    g=0.354
elif pl==7.2:
    g=0.249
elif pl==8:
    g=11.15
else:
    g=0.62


fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')

ax.set_title("3D PLOT FOR INITIAL VELOCITY, TIME AND RANGE")

t = 0.0

v = 0.0
theta = 1.0

vlist = []
thetalist = []
tlist = []

while theta < 90:
    v = 1.0
    indi = 0
    while v < 800:
        t = 2 * v * math.sin(theta * np.pi / 180) / g
        vlist.append(v)
        thetalist.append(theta)
        tlist.append(t)
        v += 1
    theta += 1

#Plotting
ax.plot3D(thetalist, vlist, tlist)

vertices = [list(zip(thetalist, vlist, tlist))]
poly = Poly3DCollection(vertices, alpha=0.8)
2 * v * math.sin(theta*np.pi/180) / g
ax.add_collection3d(poly)

#Keeping labels
ax.set_xlabel("Theta (degrees) ")
ax.set_ylabel('Velocity (m/s)')
ax.set_zlabel("Time (s) ")

plt.show()
```
